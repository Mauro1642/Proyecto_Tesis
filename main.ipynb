```python
import os
import json
from models.word_cloud import normalize_text
from models.word_cloud import generate_wordcloud
COMENTARIOS_DIR = "comentarios_por_canal"
OUTPUT_DIR = "nubes_palabras_canales"

for filename in os.listdir(COMENTARIOS_DIR):
    if filename.endswith(".json"):
        canal = filename.replace(".json", "")
        filepath = os.path.join(COMENTARIOS_DIR, filename)
        
        with open(filepath, "r", encoding="utf-8") as f:
            comentarios = json.load(f)  # Lista de strings
        
        texto_crudo = " ".join(comentarios)
        texto_limpio = normalize_text(texto_crudo)
        generate_wordcloud(texto_limpio, OUTPUT_DIR, canal)
```
