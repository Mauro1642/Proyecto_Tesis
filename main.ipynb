```python
import numpy as np
import matplotlib.pyplot as plt
import json
import networkx as nx
import pandas as pd
import os
from itertools import combinations
from datetime import datetime
import seaborn as sb
from collections import Counter
from scipy.stats import linregress
def logbin_pdf(data, nbin=100, yinf=1e-6):

    mini, maxi= min(data), max(data)

    bines = np.logspace(np.log10(mini), np.log10(maxi), nbin)
   

    counts, bines = np.histogram( data, bins=bines, density=True )

    x,y = [ bines[i]+(bines[i+1]-bines[i])*0.5 for i in range(len(bines)-1) ], counts

    x = np.array(x)
    y = np.array(y)

    x = x[y>yinf]
    y = y[y>yinf]

    return x,y
```

```python
carpeta = "comentarios_por_canal"
resultados = {}

for archivo in os.listdir(carpeta):
    if archivo.endswith(".json"):
        canal = archivo.replace("comentarios_", "").replace(".json", "")
        ruta = os.path.join(carpeta, archivo)

        with open(ruta, "r", encoding="utf-8") as f:
            data = json.load(f)

        cantidad_videos = 0
        comentarios_totales = 0
        likes_totales = 0
        views_totales = 0

        for id_video, contenido in data.items():
            if "_metrics" in contenido:
                cantidad_videos += 1

                # Likes y views del video
                likes_video = contenido["_metrics"].get("likes", 0)
                views_video = contenido["_metrics"].get("views", 0)

                if not isinstance(likes_video, (int, float)):
                    likes_video = 0
                if not isinstance(views_video, (int, float)):
                    views_video = 0

                # Contar comentarios y sumar likes de comentarios
                comentarios_video = 0
                for key, value in contenido.items():
                    if key != "_metrics" and isinstance(value, dict):
                        comentarios_video += 1
                        likes_com = value.get("likes", 0)
                        if isinstance(likes_com, (int, float)):
                            likes_video += likes_com

                comentarios_totales += comentarios_video
                likes_totales += likes_video
                views_totales += views_video

        likes_promedio = likes_totales / cantidad_videos if cantidad_videos > 0 else 0
        comentarios_promedio = comentarios_totales / cantidad_videos if cantidad_videos > 0 else 0

        resultados[canal] = {
            "cantidad_videos": cantidad_videos,
            "comentarios_totales": comentarios_totales,
            "comentarios_por_video": comentarios_promedio,
            "likes_totales": likes_totales,
            "likes_por_video": likes_promedio,
            "views_totales": views_totales
        }
```

```python
import json
import re

def limpiar_caracteres_invisibles(texto):
    # Elimina caracteres invisibles problemáticos
    return re.sub(r"[\u200B-\u200F\u202A-\u202E]", "", texto)

def eliminar_trailing_commas(texto):
    # Elimina comas antes de } o ]
    return re.sub(r",\s*([}\]])", r"\1", texto)

def cerrar_llaves(texto):
    # Cierra llaves faltantes si el archivo quedó truncado
    apertura = texto.count("{")
    cierre = texto.count("}")
    if cierre < apertura:
        texto += "}" * (apertura - cierre)
    return texto

def reparar_json_avanzado(ruta_entrada, ruta_salida):
    with open(ruta_entrada, "r", encoding="utf-8") as f:
        contenido = f.read()

    # Limpieza inicial
    contenido = limpiar_caracteres_invisibles(contenido)
    contenido = eliminar_trailing_commas(contenido)
    contenido = cerrar_llaves(contenido)

    # Intento directo
    try:
        data = json.loads(contenido)
        print("✅ JSON reparado sin necesidad de reconstrucción")
    except json.JSONDecodeError:
        print("⚠️ JSON dañado. Iniciando reparación avanzada...")

        lineas = contenido.splitlines()
        reconstruido = "{\n"
        buffer = ""

        for linea in lineas:
            buffer += linea + "\n"

            # Intentar parsear el buffer como un objeto parcial
            try:
                prueba = buffer
                prueba = eliminar_trailing_commas(prueba)
                prueba = cerrar_llaves(prueba)
                json.loads(prueba)
                reconstruido += linea + "\n"
            except:
                # Línea dañada → intentar repararla
                linea_reparada = eliminar_trailing_commas(linea)
                linea_reparada = limpiar_caracteres_invisibles(linea_reparada)
                reconstruido += linea_reparada + "\n"

        reconstruido += "}"

        # Intento final
        try:
            data = json.loads(reconstruido)
            print("✅ JSON reconstruido exitosamente")
        except Exception as e:
            print("❌ No se pudo reparar completamente el JSON")
            print("Error:", e)
            return

    # Guardar JSON limpio
    with open(ruta_salida, "w", encoding="utf-8") as f:
        json.dump(data, f, indent=2, ensure_ascii=False)

    print(f"✅ Archivo limpio guardado en: {ruta_salida}")
```

```python
import os
import json
import pandas as pd

carpeta = "analisis_por_canal"

# Lista para guardar los resultados de cada video
lista_sentimientos = []

for archivo in os.listdir(carpeta):
    nombre_canal = archivo.split("_")[1]
    ruta = os.path.join(carpeta, archivo)
    
    with open(ruta, "r", encoding="utf-8") as f:
        data = json.load(f)
    
    for video_id, comentarios in data.items():
        if not comentarios:  # si el video no tiene comentarios
            continue
        
        prom_pos = prom_neg = prom_neu = 0
        id_comentarios = list(comentarios.keys())
        
        for id_comentario in id_comentarios:
            sent_probas = comentarios[id_comentario]["sentimiento_probas"]
            prom_pos += sent_probas["POS"]
            prom_neg += sent_probas["NEG"]
            prom_neu += sent_probas["NEU"]
        
        n = len(id_comentarios)
        lista_sentimientos.append({
            "canal": nombre_canal,
            "id_video": video_id,
            "positividad": prom_pos / n,
            "negatividad": prom_neg / n,
            "neutralidad": prom_neu / n
        })

# Crear DataFrame final
df_sentimientos = pd.DataFrame(lista_sentimientos)
df_sentimientos['neutralidad+positividad']=df_sentimientos['positividad']+df_sentimientos['neutralidad']
```

```python
from manim import *
import networkx as nx
import pandas as pd
import random

class EvolucionGrafo(Scene):
    def construct(self):
        # --- Parámetros ---
        k = 100  # cantidad de videos a mostrar
        duracion_arista = 0.6  # segundos por semana
        global df_agarrapala

        # --- Procesamiento de fechas ---
        df_agarrapala['fecha'] = pd.to_datetime(df_agarrapala['fecha'])
        fecha_min = df_agarrapala['fecha'].min()
        df_agarrapala['semana'] = ((df_agarrapala['fecha'] - fecha_min).dt.days // 7) + 1

        # --- Seleccionar videos aleatorios ---
        videos_unicos = df_agarrapala['video_id'].unique()
        videos_seleccionados = random.sample(list(videos_unicos), min(k, len(videos_unicos)))
        df_agarrapala = df_agarrapala[df_agarrapala['video_id'].isin(videos_seleccionados)]

        # --- Calcular cantidad de comentarios por video ---
        comentarios_por_video = df_agarrapala.groupby("video_id")["user_id"].nunique().to_dict()
        max_comentarios = max(comentarios_por_video.values())

        # --- Calcular aristas por semana ---
        semanas = sorted(df_agarrapala['semana'].unique())
        aristas_por_semana = {}

        for semana in semanas:
            df_sem = df_agarrapala[df_agarrapala['semana'] <= semana]
            video_usuarios = df_sem.groupby('video_id')['user_id'].apply(set)
            aristas = []
            for i, v1 in enumerate(video_usuarios.index):
                for v2 in video_usuarios.index[i + 1:]:
                    inter = len(video_usuarios[v1].intersection(video_usuarios[v2]))
                    if inter > 0:
                        aristas.append((v1, v2, inter))
            aristas_por_semana[semana] = aristas

        # --- Layout estático del grafo ---
        G0 = nx.Graph()
        G0.add_nodes_from(videos_seleccionados)
        pos = nx.spring_layout(G0, seed=42, dim=2)
        pos = nx.random_layout(G0) * 2
        escala = 3

        # --- Crear nodos con tamaño/color por popularidad ---
        nodos = {}
        for n in videos_seleccionados:
            tam = 0.08 + 0.12 * np.log1p(comentarios_por_video.get(n, 1)) / np.log1p(max_comentarios)
            color = interpolate_color(BLUE, RED, comentarios_por_video.get(n, 1) / max_comentarios)
            nodo = Dot(point=[pos[n][0]*escala, pos[n][1]*escala, 0],
                       color=color, radius=tam)
            nodos[n] = nodo

        # --- Dibujar nodos y títulos ---
        self.play(*[Create(nodo) for nodo in nodos.values()], run_time=1.5)
        texto_canal = Text("Canal Agarrá la pala", font_size=20).to_corner(UL)
        texto_nodos = Text(f"{k} nodos", font_size=16).next_to(texto_canal,DOWN)
        self.play(Write(texto_nodos))
        self.play(Write(texto_canal))
        titulo = Text("Evolución de conexiones entre videos", font_size=30).to_edge(UP)
        self.play(Write(titulo))
        cuadro_leyenda = Square(side_length=0.3, color=RED).to_corner(DL)
        texto_leyenda = Text("Más comentarios", font_size=20).next_to(cuadro_leyenda, RIGHT)
        self.play(FadeIn(cuadro_leyenda), Write(texto_leyenda))

        # --- Animar aristas semana a semana ---
        aristas_existentes = set()
        total_aristas = 0  # contador acumulado

        for semana in semanas:
            nuevas_aristas = [
                (v1, v2, w) for (v1, v2, w) in aristas_por_semana[semana]
                if (v1, v2) not in aristas_existentes and (v2, v1) not in aristas_existentes
            ]
            if nuevas_aristas:
                anims = []
                for v1, v2, peso in nuevas_aristas:
                    grosor = 0.5 + 2.5 * (np.log1p(peso) / np.log1p(10))
                    color = interpolate_color(GRAY, YELLOW, min(1, peso / 10))
                    line = Line(
                        nodos[v1].get_center(),
                        nodos[v2].get_center(),
                        stroke_width=grosor,
                        color=color,
                        stroke_opacity=0.8
                    )
                    anims.append(Create(line))
                self.play(*anims, run_time=duracion_arista)
                aristas_existentes.update([(v1, v2) for v1, v2, _ in nuevas_aristas])
                total_aristas = len(aristas_existentes)  # actualizar total acumulado

            # --- Mostrar número de semana y cantidad de aristas ---
            texto_semana = Text(f"Semana {semana}", font_size=24).to_corner(UR)
            texto_aristas = Text(f"Aristas: {total_aristas}", font_size=24).next_to(texto_semana, DOWN)
            grupo_textos = VGroup(texto_semana, texto_aristas)
            self.play(Write(grupo_textos), run_time=0.4)
            self.wait(0.3)
            self.play(FadeOut(grupo_textos), run_time=0.3)

        self.wait(2)
```

```python
def graficar_evolucion_comentarios(df, video_id):
    """
    Genera gráficos de evolución de comentarios de un video y los guarda en alta calidad.

    Parámetros:
        df : pd.DataFrame
            DataFrame con columnas 'video_id' y 'fecha'
        video_id : str
            ID del video a analizar
    """
    # Asegurarse que 'fecha' sea datetime
    df['fecha'] = pd.to_datetime(df['fecha'])

    # Filtrar comentarios del video
    df_video = df[df['video_id'] == video_id].copy()
    if df_video.empty:
        print(f"No hay comentarios para el video {video_id}")
        return

    # Agrupar por día
    comentarios_por_dia = df_video.groupby(df_video['fecha'].dt.date).size().sort_index()

    # --- Gráfico 1: comentarios por día ---
    # --- Gráfico 2: comentarios acumulados ---
    acumulado = comentarios_por_dia.cumsum()
    plt.figure(figsize=(10,5))
    plt.plot(acumulado.index, acumulado.values, marker='o', color='salmon')
    plt.title(f"Comentarios acumulados - Video {video_id}")
    plt.xlabel("Fecha")
    plt.ylabel("Cantidad acumulada de comentarios")
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
```

```python
import os
import pandas as pd
import networkx as nx
import itertools
from collections import defaultdict

# === Crear carpetas de salida ===
os.makedirs('aristas_por_dia', exist_ok=True)
os.makedirs('grafos_gexf', exist_ok=True)

# === Obtener todos los videos (nodos fijos) y su canal ===
video_a_canal = dict(zip(df['video_id'], df['canal']))
todos_videos = list(video_a_canal.keys())

# === Diccionario: usuario -> videos comentados hasta el momento ===
usuarios_a_videos = defaultdict(set)

# === Diccionario final: fecha -> DataFrame de aristas ===
aristas_acumuladas = {}

# === Lista para guardar métricas ===
metricas = []

# === Recorrer fechas en orden cronológico ===
for fecha, datos_dia in df.sort_values('fecha').groupby('fecha'):
    # Actualizar comentarios acumulados
    for user, videos in datos_dia.groupby('user_id')['video_id']:
        usuarios_a_videos[user].update(videos)
    
    # Contar pares de videos con usuarios en común
    conteo_pares = defaultdict(int)
    for videos in usuarios_a_videos.values():
        for v1, v2 in itertools.combinations(sorted(videos), 2):
            conteo_pares[(v1, v2)] += 1
    
    # Crear DataFrame de aristas de esa fecha
    aristas_dia = pd.DataFrame(
        [(v1, v2, peso) for (v1, v2), peso in conteo_pares.items()],
        columns=['video_origen', 'video_destino', 'peso']
    )
    
    # Guardar en diccionario
    aristas_acumuladas[fecha] = aristas_dia

    # === Exportar CSV de aristas ===
    nombre = fecha.strftime('%Y-%m-%d')
    aristas_dia.to_csv(f'aristas_por_dia/aristas_{nombre}.csv', index=False)

    # === Crear grafo ===
    G = nx.Graph()
    
    # Agregar nodos con su atributo "canal"
    for v in todos_videos:
        canal = video_a_canal.get(v, "Desconocido")
        G.add_node(v, canal=canal)
    
    # Agregar aristas con peso
    for _, fila in aristas_dia.iterrows():
        G.add_edge(fila['video_origen'], fila['video_destino'], weight=fila['peso'])

    # === Calcular métricas ===
    n_nodos = G.number_of_nodes()
    n_aristas = G.number_of_edges()
    densidad = nx.density(G)
    n_componentes = nx.number_connected_components(G)

    # Componente gigante
    if n_aristas > 0 and n_componentes > 0:
        comp_gigante = max(nx.connected_components(G), key=len)
        Gg = G.subgraph(comp_gigante)
        n_nodos_gigante = Gg.number_of_nodes()
        n_aristas_gigante = Gg.number_of_edges()
        densidad_gigante = nx.density(Gg)
    else:
        n_nodos_gigante = 0
        n_aristas_gigante = 0
        densidad_gigante = 0

    # Guardar métricas
    metricas.append({
        'fecha': nombre,
        'n_nodos_total': n_nodos,
        'n_aristas_total': n_aristas,
        'densidad_total': densidad,
        'n_componentes': n_componentes,
        'n_nodos_gigante': n_nodos_gigante,
        'n_aristas_gigante': n_aristas_gigante,
        'densidad_gigante': densidad_gigante
    })

    # === Exportar GEXF (para Gephi) ===
    nx.write_gexf(G, f'grafos_gexf/grafo_{nombre}.gexf')

    print(f"{nombre}: {n_nodos} nodos, {n_aristas} aristas, {n_componentes} componentes. "
          f"Gigante: {n_nodos_gigante} nodos, {n_aristas_gigante} aristas.")

# === Guardar métricas en CSV ===
df_metricas = pd.DataFrame(metricas)
df_metricas.to_csv('metricas_por_dia.csv', index=False)

print("\n✅ Proceso completo. Todos los grafos, aristas y métricas guardados correctamente.")
```

```python
from networkx.algorithms.community import louvain_communities
from scipy.cluster.hierarchy import dendrogram, linkage
# === Cargar datos ===
edges = pd.read_csv("edgescanales.csv")
nodes = pd.read_csv("nodescanales.csv")

# === Construir grafo no dirigido ===
G = nx.from_pandas_edgelist(edges, source='Source', target='Target', edge_attr='Weight')

# === Agregar atributos de los nodos ===
for _, row in nodes.iterrows():
    if row['Id'] in G.nodes:
        G.nodes[row['Id']]['Label'] = row['Label']
        G.nodes[row['Id']]['num_videos'] = row['num_videos']
        G.nodes[row['Id']]['total_usuarios'] = row['total_usuarios']

# === Detección de comunidades con Louvain (implementación de NetworkX) ===

comunidades = louvain_communities(G, weight='Weight', seed=42)
Q = modularity(G, comunidades)
print(Q)

# Crear un diccionario {nodo: comunidad_id}
partition = {}
for i, comunidad in enumerate(comunidades):
    for nodo in comunidad:
        partition[nodo] = i
nx.set_node_attributes(G, partition, "comunidad")

# === Visualización ===
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G, seed=42)
colors = [partition[n] for n in G.nodes()]
nx.draw_networkx_nodes(G, pos, node_color=colors, cmap=plt.cm.Set3, node_size=200)
nx.draw_networkx_edges(G, pos, alpha=0.3)
nx.draw_networkx_labels(G, pos, {n: G.nodes[n]['Label'] for n in G.nodes()}, font_size=8)
plt.title("Comunidades detectadas (Louvain - NetworkX)")
plt.axis("off")
plt.tight_layout()
plt.show()

nx.write_gexf(G, "canales_con_comunidades.gexf")
```

```python
from matplotlib import cm
# === Matriz de adyacencia ponderada ===
A = nx.to_pandas_adjacency(G, weight="Weight")

# === Calcular linkage jerárquico (Ward minimiza la varianza intra-grupo) ===
Z = linkage(A, method="ward")

# === Etiquetas y comunidades ===
labels = [G.nodes[n]["Label"] for n in A.index]
comunidades = [partition[n] for n in A.index]

# === Mapa de colores para comunidades ===
unique_coms = sorted(set(comunidades))
cmap = cm.get_cmap("tab20", len(unique_coms))
colors = [cmap(c) for c in comunidades]

# === Graficar dendrograma con colores ===
plt.figure(figsize=(14, 6))
ddata = dendrogram(Z, labels=labels, leaf_rotation=90, leaf_font_size=9)

# === Colorear etiquetas según comunidad ===
ax = plt.gca()
xlbls = ax.get_xmajorticklabels()
for lbl in xlbls:
    canal = lbl.get_text()
    idx = labels.index(canal)
    lbl.set_color(colors[idx])

plt.title("Dendrograma de canales según similitud estructural (coloreado por comunidad Louvain)")
plt.xlabel("Canales")
plt.ylabel("Distancia")
plt.xticks(fontsize=14)
plt.tight_layout()
plt.show()
```

```python
import re

# Estados posibles
ROOT = "root"
VIDEO = "video"
COMMENT = "comment"
METRICS = "metrics"

state_stack = []
errors = []

def current():
    return state_stack[-1] if state_stack else None

with open(archivo, encoding="utf-8", errors="ignore") as f:
    for lineno, line in enumerate(f, start=1):

        stripped = line.strip()

        # Detectar apertura de objeto raíz
        if stripped == "{":
            if state_stack:
                state_stack.append("object")
            else:
                state_stack.append(ROOT)
            continue

        # Detectar cierre
        if stripped in ("}", "},"):
            if not state_stack:
                errors.append((lineno, "Cierre } sin apertura"))
            else:
                state_stack.pop()
            continue

        # Detectar clave
        m = re.match(r'"([^"]+)"\s*:\s*{', stripped)
        if m:
            key = m.group(1)

            if current() == ROOT:
                state_stack.append(VIDEO)
            elif current() == VIDEO:
                if key == "_metrics":
                    state_stack.append(METRICS)
                else:
                    state_stack.append(COMMENT)
            else:
                errors.append((lineno, f"Clave '{key}' en nivel inválido"))
            continue

        # Detectar apertura de objeto genérico
        if stripped.endswith("{"):
            state_stack.append("object")
            continue

# Resultados
print("📊 Niveles abiertos al final:", state_stack)

if errors:
    print("\n❌ ERRORES ESTRUCTURALES:")
    for e in errors:
        print("Línea", e[0], "-", e[1])
else:
    print("\n✅ No se detectaron errores estructurales")
```

```python
import re

def detectar_video_mal_cerrado(path):
    inicio_video_re = re.compile(r'^\s*"([^"]+)"\s*:\s*\{')
    
    video_actual = None
    linea_inicio_video = None
    balance = 0
    errores = []

    with open(path, "r", encoding="utf-8") as f:
        for num_linea, linea in enumerate(f, 1):

            # ¿Arranca un nuevo video?
            m = inicio_video_re.match(linea)
            if m:
                # Si ya había un video abierto y no cerró
                if video_actual is not None and balance != 0:
                    errores.append({
                        "video": video_actual,
                        "linea_inicio": linea_inicio_video,
                        "balance_restante": balance,
                        "linea_error": num_linea
                    })

                video_actual = m.group(1)
                linea_inicio_video = num_linea
                balance = 0

            # Contar llaves SOLO si estamos dentro de un video
            if video_actual is not None:
                balance += linea.count("{")
                balance -= linea.count("}")

                # Video correctamente cerrado
                if balance == 0:
                    video_actual = None
                    linea_inicio_video = None

    # Si el archivo terminó con un video abierto
    if video_actual is not None:
        errores.append({
            "video": video_actual,
            "linea_inicio": linea_inicio_video,
            "balance_restante": balance,
            "linea_error": "EOF"
        })

    return errores
```

```python
errores = detectar_video_mal_cerrado(archivo)

if not errores:
    print("✅ No se detectaron videos mal cerrados")
else:
    print("❌ Videos con problemas de cierre:")
    for e in errores:
        print(
            f"- Video: {e['video']} | "
            f"Empieza en línea {e['linea_inicio']} | "
            f"Balance restante: {e['balance_restante']} | "
            f"Detectado en línea {e['linea_error']}"
        )
```

```python


with open(archivo, encoding="utf-8", errors="ignore") as f:
    lines = f.readlines()

video_starts = []
video_ends = []

# 1️⃣ detectar inicios y cierres de videos
for i, line in enumerate(lines):
    stripped = line.rstrip()

    # inicio de video: dos espacios + "algo":{
    if stripped.startswith('  "') and stripped.endswith('": {'):
        video_starts.append(i + 1)

    # cierre de video: exactamente dos espacios + }
    if stripped == "  }" or stripped == "  },":
        video_ends.append((i + 1, stripped.endswith(",")))

# 2️⃣ chequeos
print("📊 Videos detectados:", len(video_starts))
print("📊 Cierres detectados:", len(video_ends))

if len(video_starts) != len(video_ends):
    print("❌ Cantidad distinta de inicios y cierres de video")
else:
    print("✅ Cantidad de inicios y cierres coincide")

# 3️⃣ verificar comas
errores = []

for idx, (linea, tiene_coma) in enumerate(video_ends):
    es_ultimo = (idx == len(video_ends) - 1)

    if not es_ultimo and not tiene_coma:
        errores.append((linea, "Falta coma en cierre de video"))
    if es_ultimo and tiene_coma:
        errores.append((linea, "Último video NO debe tener coma"))

# 4️⃣ reporte
if errores:
    print("❌ Errores encontrados:")
    for e in errores[:10]:
        print(e)
else:
    print("✅ Todos los videos están correctamente separados")
```

```python


with open(archivo, encoding="utf-8", errors="ignore") as f:
    lines = f.readlines()

stack = []   # pila de videos abiertos
errores = []

for i, line in enumerate(lines):
    stripped = line.rstrip()

    # apertura de video
    if stripped.startswith('  "') and stripped.endswith('": {'):
        video_id = stripped.split('"')[1]
        stack.append((video_id, i + 1))

    # cierre de video
    if stripped == "  }" or stripped == "  },":
        if stack:
            stack.pop()
        else:
            errores.append(("Cierre sin apertura", i + 1))

print("Videos sin cerrar:", stack[:5])
print("Cantidad:", len(stack))
```
